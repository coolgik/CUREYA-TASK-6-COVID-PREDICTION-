**DATA SPLITTING:**

```python
from sklearn.model_selection import train_test_split
```

```python
#converting string Observationdate to datetime
import datetime as dt
india_case["ObservationDate"] = pd.to_datetime(india_case["ObservationDate"])
india_case.head()
```

```python
#converting datetime to ordinal
india_case["ObservationDate"] = india_case["ObservationDate"].map(dt.datetime.toordinal)
```

```python
#getting dependent and independent variable
x = india_case["ObservationDate"]
y = india_case["Confirmed"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=0)
```

**LOGISTIC REGRESSION:**

```python
from sklearn.linear_model import LogisticRegression
```

```python
logReg = LogisticRegression()
```

```python
import numpy as np
logReg.fit(np.array(x_train).reshape(-1,1),np.array(y_train).reshape(-1,1))
```

```python
india_case.tail()
```

```python
india_case.head()
```

```python
y_pred = logReg.predict(np.array(x_test).reshape(-1,1))
```

```python
from sklearn.metrics import confusion_matrix, classification_report,accuracy_score,mean_squared_error
```

```python
confusion_matrix(x_test,y_pred)
```

```python
classification_report(x_test,y_pred)
```

```python
accuracy_score(x_test,y_pred)
```

```python
mean_squared_error(x_test,y_pred)
```

```python
cm = confusion_matrix(y_test,y_pred)
```

```python
cm
```

```python
plt.figure(figsize=(10,13))
plt.scatter(y_test, y_pred)
plt.plot([y_test.min(),y_test.max()],[y_test.min(),y_test.max()], color='blue', linewidth=3)
plt.xlabel("y_test")
plt.ylabel("y_pred")
plt.title("y_test vs y_pred")
```
